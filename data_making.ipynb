### 2. create train and test folders

```python
#train

df_train = pd.read_pickle('flowers/text2ImgData.pkl')
```

```python
class_map = {}
```

```python
for i, txt_file in enumerate(sorted(glob('flowers/flowers/images/*'))):
    class_map[i + 1] = txt_file[-11:]
```

```python
base = 'flowers/flowers/images'
os.path.join(base, class_map[2])
```

```python
pk_img_classes = []
pk_img_paths = []
for i in range(df_train.shape[0]):
    img = df_train.iloc[i]['ImagePath'][-15:-4]
    img_id = df_train.index[i]
    img_class = img_labels[img_id - 1]
    img_path = os.path.join(class_map[img_class],(img))
    pk_img_paths.append(img_path)
    pk_img_classes.append(img_class)
```

### 3. captions pickle

```python
pk_img_paths
```

```python
with open('flowers/vocab.pkl','rb') as f:
    wordtoix = pickle.load(f)
```

```python
# len(wordtoix)
# ixtoword[5446]
index = len(wordtoix) + 1
```

#### 3.1 train

```python
train_text_encodings = []
for i, img_path in enumerate(pk_img_paths):
    text_file_path = os.path.join('flowers/text',(img_path + '.txt'))
    with open(text_file_path,'r') as file:
        for line in file:
            sent = re.sub(r'[^\w\s]', ' ', line)
            words = sent.split()
            word_idx = []
            for word in words:
                if word not in wordtoix.keys():
                    wordtoix[word] = index
                    index += 1
                word_idx.append(wordtoix[word])                
            train_text_encodings.append(word_idx)
```

```python
train_text_encodings[0]
```

```python
test_text_encodings = []
for i, img_path in enumerate(test_img_paths):
    text_file_path = os.path.join('flowers/text',(img_path + '.txt'))
    with open(text_file_path,'r') as file:
        for line in file:
            sent = re.sub(r'[^\w\s]', ' ', line)
            words = sent.split()
            word_idx = []
            for word in words:
                if word not in wordtoix.keys():
                    wordtoix[word] = index
                    index += 1
                word_idx.append(wordtoix[word])                
            test_text_encodings.append(word_idx)
```

```python
len(test_text_encodings)
```

```python
len(wordtoix)
```

```python
ixtoword = {}
for key, value in wordtoix.items():
    ixtoword[value] = key
```

```python
len(ixtoword)
```

```python
captions = [train_text_encodings, test_text_encodings, ixtoword,wordtoix]
```

```python
with open('flowers/captions.pickle','wb') as f:
    pickle.dump(captions, f)
```
